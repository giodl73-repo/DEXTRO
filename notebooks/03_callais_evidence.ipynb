# Load the bloc_voting.json output from a prior `redist analyze` run.
import json
with open('outputs/v1/2020/plans/la_2020_callais/analysis/bloc_voting.json') as f:
    bv = json.load(f)

# Print the headline regression specification (the SCALE-block-lifting receipt).
print(bv['candidates'][0]['regression']['specification'])
# Expected: 'WLS weighted by precinct vote count; HC3 robust SE; cluster-bootstrap by county B=...'

# Per-candidate race coefficient + 95% cluster CI, with Holm-corrected p.
for c in bv['candidates']:
    coef = c['regression']['coefficients']['pct_minority']
    print(f"{c['candidate']}: beta={coef['estimate']:.3f}, "
          f"95% CI [{coef['ci_95_cluster'][0]:.3f}, {coef['ci_95_cluster'][1]:.3f}], "
          f"p_holm={coef['p_value_holm']:.4f}")
    print(f"  ecology caveat: {c['ecology_caveat']}")
    print(f"  draft interpretation: {c['draft_interpretation']}")

# Verify race-of-candidate provenance chain.
prov = bv['race_of_candidate_provenance']
print(f"\nProvenance: {prov['source_file']} sha256={prov['source_sha256'][:16]}...")
print(f"Independently verified: {prov['annotations_independently_verified']}")
for c in prov['curators']:
    print(f"  Curator: {c['curator']} ({c['curator_credentials']}), {c['n_candidates']} candidates")
