# 04 — GerryChain Interop

**Persona:** academic comparing `redist` to GerryChain (the field's standard MCMC library).
**Time:** ≤ 120 seconds.
**Outcome:** demonstrate the round-trip property `T(redist(T(P))) == P` modulo district-label permutation, where T is GerryChain's `Partition.to_json()`.

This is the notebook that proves `redist` and GerryChain agree on what a redistricting plan IS — they encode the same partition, just in slightly different file formats. The round-trip property test (Researcher Toolkit plan Task 4) lives here.

**Requires:** `gerrychain` installed in the same Python environment. The kernel-state attestation cell will SKIP this notebook if GerryChain isn't present.


````python
from packaging.specifiers import SpecifierSet
REDIST_PY_RANGE = '>=0.4,<0.5'
GERRYCHAIN_RANGE = '>=0.3.2,<0.4'

import redist_py
assert redist_py.__version__ in SpecifierSet(REDIST_PY_RANGE)

try:
    import gerrychain
    assert gerrychain.__version__ in SpecifierSet(GERRYCHAIN_RANGE), \
        f'gerrychain {gerrychain.__version__} not in compatible range {GERRYCHAIN_RANGE}'
    print(f'redist_py {redist_py.__version__}, gerrychain {gerrychain.__version__} OK')
except ImportError:
    print('SKIP: gerrychain not installed; this notebook requires it')
    print('Install: pip install gerrychain==0.3.5')
    raise SystemExit(0)
````
