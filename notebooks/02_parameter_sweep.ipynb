```python
import pandas as pd
import itertools

results = []
seeds = range(20)
tolerances = [0.001, 0.005, 0.01]
for seed, tol in itertools.product(seeds, tolerances):
    plan = redist_py.bisect('VT', 2020, n_districts=1, balance_tolerance=tol, seed=seed)
    results.append({
        'seed': seed,
        'tolerance': tol,
        'mean_pp': plan.mean_polsby_popper(),
        'pop_max_dev': plan.max_population_deviation(),
    })
df = pd.DataFrame(results)
df.groupby('tolerance').agg({'mean_pp': ['mean', 'std']})
```
